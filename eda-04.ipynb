```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical columns and compute correlation matrix
numerical_df = df[['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score']]
corr_matrix = numerical_df.corr()

# Plot heatmap
sns.heatmap(data=corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Variables")
plt.show()
```

```python
# Create bins for sleep hours (example: 3-4, 4-5, etc.)
df['Sleep_Bins'] = pd.cut(df['Sleep_Hours_Per_Night'], bins=5)

# Pivot table: Mean usage hours per sleep bin
pivot_table = df.pivot_table(values='Avg_Daily_Usage_Hours', 
                             index='Sleep_Bins', 
                             aggfunc='mean')

sns.heatmap(data=pivot_table, annot=True, fmt='.1f', cmap='YlGnBu')
plt.title("Avg Social Media Usage by Sleep Hours (Binned)")
plt.show()
```
